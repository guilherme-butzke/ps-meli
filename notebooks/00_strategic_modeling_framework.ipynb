# Strategic Modeling Framework: Profit-Optimized Fraud Detection

Request: 
> Train an algorithm with the objective of predicting fraud, considering that for each transaction the profit margin is 25%, and that for each approved fraudulent transaction, 100% of the transaction value is lost. Perform an analysis and determine a model that maximizes the company’s profit.

To model this problem, we must shift from standard accuracy-based metrics to a **Financial Cost Function**. The goal is to find the threshold where the cost of a "False Alarm" (blocking a good customer) equals the cost of a "Missed Fraud" (allowing a thief to transact).

The goal of this phase is to define the profit objective function to optimize the profit, and create a framework to solve this problem as a ML approach. 

## 1. The Profit Objective Function

The goal of this analysis is:

> *"Improve profit by reducing loss using a model able to detect fraud."*

If our model is to detect a fraud as a `positive` and a legitimate transaction as a `negative`, we must understand the outcomes of our model in terms of business actions:

* **True Positive (TP):** Blocked fraudulent transaction.
* **False Positive (FP):** Blocked legitimate transaction.
* **True Negative (TN):** Approved legitimate transaction.
* **False Negative (FN):** Approved fraudulent transaction.

Also, we must recognize that the model's mistakes are our true costs. Both False Negatives ($FN$) and False Positives ($FP$) represent financial losses to the business:

* **False Negatives ($FN$):** The direct loss of the total transaction value to fraud (Approved Fraud).
* **False Positives ($FP$):** The opportunity cost of losing our 25% profit margin on blocked legitimate customers (Blocked Legitimate).

Conversely, True Positives ($TP$) and True Negatives ($TN$) do not actively generate *new* profit from the perspective of the model's action. As a guiding principle:

> *"Preventing a loss is not the same as generating revenue."*

* **True Positive ($TP$):** Successfully blocked fraud. The financial loss is avoided, but no new profit is generated.
* **True Negative ($TN$):** Successfully approved legitimate transaction. The baseline revenue is generated, but this is unaffected by the model's intervention.

Therefore, instead of trying to maximize total revenue directly, we need to minimize the financial damage caused by the model's errors. We optimize the following cost function:

$$M = - (fraud\_loss \cdot FN) - (margin \cdot FP)$$

or 

$$M = - (1.0 \cdot FN) - (0.25 \cdot FP)$$

We must to highlight the ideia here that: 

> For fraud prevention, maximizing the company’s profits does not mean being profitable, but reducing losses.

**Mathematical Proof**

We can prove that maximizing our optimization metric ($M$) is mathematically equivalent to maximizing the absolute profit. Here, $fraud\_loss$ is declared as 100% of the value of the transaction, and $margin$ is declared as 25% of the transaction value (the amount lost when a good transaction does not complete).

The absolute profit equation is:

$$M_{absolute} = (margin \cdot TN) - (fraud\_loss \cdot FN)$$

We know that the total transaction value of all legitimate customers is the sum of True Negatives and False Positives ($Total\_Legit = TN + FP$). Therefore, we can substitute $TN = Total\_Legit - FP$:

$$M_{absolute} = margin \cdot (Total\_Legit - FP) - (fraud\_loss \cdot FN)$$

Expanding this equation gives us:

$$M_{absolute} = \underbrace{(margin \cdot Total\_Legit)}_{\text{Constant } (C)} - (margin \cdot FP) - (fraud\_loss \cdot FN)$$


Because the total possible margin from all legitimate customers is a fixed constant ($C$) outside of the model's control, the equation simplifies to:

$$M_{absolute} = C + M$$

Since $C$ is a constant, maximizing our cost function $M$ perfectly maximizes the absolute profit of the company.

## 2. Development Strategy

To construct this model to predict fraudulent transactions, we followed an iterative approach close to the CRISP methodology. Each notebook documents specific actions, analyzes results, and refines the strategy for the subsequent phase.

In the ``notebooks`` folder, each notebook file, beyond the code used to generate the results, documents specific actions, analyzes results, and refines the strategy for the subsequent phase.

These are the steps:


### Phase 1: Exploratory Data Analysis (EDA)
* **Dataset Description:** Identify the target variable (`fraude`) and analyze its behavior and relationships with other features in the dataset.
* **Insight Extraction:** Identify statistical anomalies within variables, such as high-frequency categories or unusual temporal patterns.
* **Feature Selection:** Based on the EDA, determine which variables to keep, which to drop immediately, and identify opportunities for new features derived from the data distribution.

### Phase 2: Feature Engineering & Target Analysis 

* **Technical Literature Review:** Research and select specific preprocessing techniques (e.g., handling extreme skewness, high cardinality, and missingness) identified during the Phase 1 exploration.
* **Feature Prioritization:** Compile a list of potential new variables and rank them by their expected predictive power and implementation complexity.
* **Feature Engineering:** Derive new features based on the initial hypotheses (such as temporal extractions from `fecha` or behavioral ratios) to enhance the model's predictive performance.
* **Target Characterization:** Analyze the distribution of the `fraude` variable and its specific relationship with other variables.
* **Variable Relevance:** Use statistical tests (e.g., Chi-Squared for categorical data, ANOVA for numerical data) to quantify the relationship between features and the target.
* **Hypothesis Formulation:** Create a list of "Risk Indicators" (e.g., "Transactions above amount $X$ are $Y\%$ more likely to be fraud").
* **Feature Selection:** If necessary, other feature selection could be done after this previous steps. 

### Phase 3: Model Selection & Profit Optimization
* **Algorithm Candidates:** Train high-performance classifiers (e.g., Random Forest, XGBoost) and establish a baseline using Logistic Regression.
* **Cost-Sensitive Learning:** Incorporate the 4:1 loss ratio ($100\%$ loss vs. $25\%$ margin) directly into the model training weights or cost function.
* **Threshold Optimization:** Instead of using the default 0.5 threshold, identify the specific probability cutoff (theoretically **0.20**) that maximizes the Profit Function.

**Iterative Refinement:** At the conclusion of each phase, the strategy will be updated based on data findings. The final output will be a production-ready ML model optimized for **Maximum Net Profit**.
